# Data Analysis

```python
import numpy as np
import json
import cv2
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
with open('data/train_data.json', 'r') as f:
    data = json.load(f)
```

```python
files = list(data.keys())
```

```python
IDX = 1000
```

```python
image = cv2.imread(files[IDX])
files[IDX]
```

```python
image.shape
```

```python
coord = data[files[IDX]]
coord
```

```python
x, y, w, h = coord['x'], coord['y'], coord['w'], coord['h']
```

```python
cv2.circle(image, (x, y),1,(0,255,0),2)
cv2.rectangle(image, (x, y), ((x + w), (y + h)), (0, 255, 0), 2)
plt.axis("off")
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```
